```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
plt.scatter(BESS_df.insitu_LE_Wm2, JET_df.insitu_LE_Wm2, alpha=0.5)
plt.xlabel('BESS_df.insitu_LE_Wm2')
plt.ylabel('JET_df.insitu_LE_Wm2')
plt.title('JET vs BESS In-situ Latent Heat Flux')
plt.plot([BESS_df.insitu_LE_Wm2.min(), BESS_df.insitu_LE_Wm2.max()], 
         [BESS_df.insitu_LE_Wm2.min(), BESS_df.insitu_LE_Wm2.max()], 
         'r--', label='1:1 line')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
