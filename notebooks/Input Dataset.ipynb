```python
from os import getcwd
from os.path import join, dirname
from FLiESANN import load_ECOv002_calval_FLiESANN_inputs
from BESS_JPL import load_ECOv002_calval_BESS_inputs
from STIC_JPL import load_ECOv002_calval_STIC_inputs
from PMJPL import load_ECOv002_calval_PMJPL_inputs
from PTJPLSM import load_ECOv002_calval_PTJPLSM_inputs
from ECOv003_L3T_L4T_JET import process_JET_table
```
